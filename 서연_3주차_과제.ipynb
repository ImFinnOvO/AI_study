```python
def print_n_times(value, i):
  for i in range(i):
    print(value)

print_n_times("안녕하세요", 5)
```

```python
def print_n_times(n, *values):
  for i in range(n):
    for value in values:
      print(value)
    print()

print_n_times(3, '안녕하세요','즐거운','파이썬 프로그래밍')
```

```python
def sum_all(start, end):
  output = 0
  for i in range(start, end+1):
    output += i
  return output

print(sum_all(0,100))
```

```python
def sum_all(start, end, step=1):
  output = 0
  for i in range(start, end+1,step):
    output += i
  return output

print(sum_all(0,100,2))
```

```python
def factorial(n):
  output = 1
  for i in range(1, n+1):
    output *= i
  return output

factorial(4)
```

```python
def factorail(n):
  if n == 0:
    return 1
  else:
    return n * factorail(n - 1)

print(factorial(4))
```

```python
def fibonacci(n):
  if n == 1:
    return 1
  if n == 2:
    return 1
  else:
    return fibonacci(n - 1) + fibonacci(n - 2)

fibonacci(10)
```

```python
counter = 0
def fibonacci(n):
  print('fibonacci({})를 구합니다.'.format(n))
  global counter
  counter += 1
  if n == 1:
    return 1
  if n == 2:
    return 1
  else:
    return fibonacci(n-1)+fibonacci(n-2)

fibonacci(35)
```

```python
dictionary = {
    1:1,
    2:1
}

def fibonacci(n):
  if n in dictionary:
    return dictionary[n]
  else:
    output = fibonacci(n -1)+ fibonacci(n- 2)
    dictionary[n] = output
    return output

fibonacci(50)
```

```python
def power(item):
  return item*item
def under_3(item):
  return item < 3

list_input_a = [1, 2, 3, 4, 5]

output_a = map(power,list_input_a)
print(output_a)
print(list(output_a))

output_b = filter(under_3, list_input_a)
print(output_b)
print(list(output_b))
```

```python
with open('/content/drive/MyDrive/info.txt','r',encoding="CP949") as file:
  for line in file:
    (name, weight, height) = line.strip().split(",")
    if (not name) or (not weight) or (not height):
      continue
    bmi = int(weight) / ((int(height)/100)**2)
    result = ''
    if 25 <= bmi:
      result = '과체중'
    elif 19.5 <= bmi:
      result = '정상 체중'
    else:
      result = '저체중'
    
    print('\n'.join([
                     '이름: {}'
                     '몸무게: {}'
                     '키: {}'
                     'BMI: {}'
                     '결과: {}'
    ])
                     .format(name,weight,height,bmi,result))
```
